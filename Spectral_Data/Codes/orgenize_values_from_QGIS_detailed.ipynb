# make the table:

read the file:

```python
data = pd.read_csv(path)
data
```

transpose the table:

```python
data = data.T
```

remove unnecessarry row:

```python
if 'id' in data.index:
  data.drop ('id', inplace = True)
```

add column for the names of the bands and split in into 2:

```python
data['long_name'] = data.index
data['long_name'] = data['long_name'].str.slice(start=1)
data[['band','stat']] = data.long_name.str.split("_",expand=True)
```

make the "band" column to integers:

```python
data['band'] = data['band'].str.slice(start=1)
data = data.astype({'band':'int'})
```

now we fix the names of the categories that sliced in the extraction:

```python
num = 0
for word in data['stat']:
  if word == 'stde':
    data['stat'].iloc[num] = 'stdev'
  if word == 'medi' or word == 'media':
    data['stat'].iloc[num] = 'median'
  num += 1
```

now we sort the table by names of the categories and the bands:

```python
data = data.sort_values(['stat','band'])
```

now we can set the bands as the index:

```python
data = data.set_index('band')
```

lets create all the needed columns and transfer the data in:

```python
list = ['wavelength','mean','stdev','stdev/mean','stdev','median','min','max']
for i in list:
  data[i] = 0
```

```python
#its a better idea and i should think about it
###data['mean'].iloc[0:274] = data[data['stat']=='mean']
```

```python
data['max'].iloc[0:274] = (data[0].iloc[0:274])
data['mean'].iloc[0:274] = data[0].iloc[274:548]
data['median'].iloc[0:274] = data[0].iloc[548:822]
data['min'].iloc[0:274] = data[0].iloc[822:1096]
data['stdev'].iloc[0:274] = data[0].iloc[1096:1370]
data['wavelength'].iloc[0:274] = wavelength_col[0:274]
data['stdev/mean'] = data['stdev'] / data['mean']
```

lets drop unnecessary columns and rows:

```python
lst = ['long_name','stat',0]
for i in lst:
  data = data.drop (i, axis = 1)
```
