# central table - mean

```python
central_table = pd.read_csv('/content/drive/MyDrive/projects/amirim/central_data_table_2.csv')
central_table = central_table.drop('Unnamed: 0', axis = 1)
```

```python
dir = '/content/drive/MyDrive/projects/amirim/VI_values/12.22/orgenized/'
trees_list = glob.glob(dir+'*.csv')
print(len(trees_list))
trees_list
```

## If it's new data that we don't yet for this month:

```python
for tree_table in trees_list:
  
  source_table = pd.read_csv(tree_table)
  
  #create new dataframe:
  df = pd.DataFrame()

  #date
  date = '20-12-22'
  df.loc[0, 'date'] = datetime.datetime.strptime(date, '%d-%m-%y').date()

  #names and plots headers
  short_names = ['cera', 'pine', 'pist', 'quer', 'cyp']
  names_dict = {'cera':('carob','Ceratonia siliqua', 'Csi'),'pine':('pine', 'Pinus halepensis', 'Ph'), 'pist':('pistacia', 'Pistacia lentiscus', 'Pl'), 'quer':('oak', 'Quercus calliprinos', 'Qc'), 'cyp':('cypress', 'Cupressus sempervirens', 'Cse')}
  treatment_dict = {'1':'control', '2':'drought', '3':'control', '4':'control', '5':'drought', '6':'drought'}
  plot_number = tree_table[-5]
  tree_short = tree_table[65:-6]  #for file names without date
  #tree_short = tree_table[71:-6] #for file names with date

  df.loc[0, 'short_name'] = tree_table[65:-4]  #for file names without date
  #df.loc[0, 'short_name'] = tree_table[71:-4]  #for file names with date
  df.loc[0, 'English_name'] = names_dict[tree_short][0]
  df.loc[0, 'Latin_name'] = names_dict[tree_short][1]
  df.loc[0, 'plot'] = plot_number 
  df.loc[0, 'treatment'] = treatment_dict[plot_number]
  df.loc[0, 'ID'] = names_dict[tree_short][2] +plot_number

  #add LWP and LAI columns
  df.loc[0, 'LWP'] =''
  df.loc[0, 'LAI'] =''

  #add the spectral bands
  cols = [str(band) for band in source_table['band']]
  df.loc[0, cols] = source_table['mean'].tolist() 

  #add the new df to the central table:
  central_table = central_table.append(df)
```

## If we have LWP data in this month:

```python
#change before any time!
date = '2022-12-20'

for tree in trees_list:
  short_name = tree[65:-4]
  tree_table = pd.read_csv(tree)
  row_index = (central_table['short_name'] == short_name) & (central_table['LWP_date'] == date) #this could be changed if it is not the same date!
  # get the mean values for this tree and date from the tree_table dataframe
  mean_values = tree_table['mean']
  central_table.loc[row_index, 'spectral_date'] = date
  central_table.loc[row_index, '1':'274'] = mean_values.values

 # for i, row in tree_table.iterrows():
    
```
